```python
# ============================================================
# INSTALLATION
# ============================================================
!pip install -q drain3 cryptography pandas numpy scikit-learn torch tqdm matplotlib

# ============================================================
# IMPORTS
# ============================================================
import os
import hashlib
import sqlite3
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import torch
import torch.nn as nn

from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score

from drain3 import TemplateMiner
from drain3.template_miner_config import TemplateMinerConfig

from cryptography.hazmat.primitives.asymmetric.ed25519 import Ed25519PrivateKey

# ============================================================
# CONFIG
# ============================================================
df_logs = pd.read_csv(BGL_FILE)
y_true = df_logs["Label"]
BGL_FILE = "/content/drive/MyDrive/BGL_2k.log_structured.csv"
SQLITE_DB = "logs_security.db"
MAX_LOGS = 5000
SEQ_LEN = 10

# ============================================================
# SQLITE
# ============================================================
conn = sqlite3.connect(SQLITE_DB)
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS parsed_logs (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    original_log TEXT,
    template TEXT,
    cluster_id INTEGER,
    anomaly INTEGER,
    signature TEXT,
    merkle_leaf TEXT,
    timestamp DATETIME DEFAULT CURRENT_TIMESTAMP
)
""")
conn.commit()

print("✅ SQLite initialisée")

# ============================================================
# DRAIN3
# ============================================================
config = TemplateMinerConfig()
config.drain_depth = 4
config.drain_sim_th = 0.4
config.drain_max_children = 100

template_miner = TemplateMiner(config=config)
print("✅ Drain initialisé")

# ============================================================
# MERKLE TREE SIMPLE
# ============================================================
class SimpleMerkleTree:
    def __init__(self):
        self.leaves = []

    def add(self, data: bytes):
        self.leaves.append(hashlib.sha256(data).hexdigest())

    def root(self):
        nodes = self.leaves.copy()

        if not nodes:
            return None

        while len(nodes) > 1:
            temp = []
            for i in range(0, len(nodes), 2):
                left = nodes[i]
                right = nodes[i+1] if i+1 < len(nodes) else left
                temp.append(
                    hashlib.sha256((left + right).encode()).hexdigest()
                )
            nodes = temp

        return nodes[0]

tree = SimpleMerkleTree()

# ============================================================
# ED25519
# ============================================================
private_key = Ed25519PrivateKey.generate()
public_key = private_key.public_key()

def sign_data(data):
    return private_key.sign(data.encode()).hex()

def verify_signature(data, sig):
    try:
        public_key.verify(bytes.fromhex(sig), data.encode())
        return True
    except:
        return False

# ============================================================
# LOAD LOGS
# ============================================================
logs = []

with open(BGL_FILE, "r", errors="ignore") as f:
    for i, line in enumerate(f):
        if i >= MAX_LOGS:
            break
        logs.append(line.strip())

print("✅ Logs chargés :", len(logs))

# ============================================================
# DRAIN PARSING
# ============================================================
templates, cluster_ids = [], []

for log in logs:
    try:
        res = template_miner.add_log_message(log)
        templates.append(res["template_mined"])
        cluster_ids.append(res["cluster_id"])
    except:
        templates.append("UNKNOWN")
        cluster_ids.append(-1)

print("✅ Parsing terminé")

# ============================================================
# ENCODING
# ============================================================
encoder = LabelEncoder()
encoded = encoder.fit_transform(templates)

# ============================================================
# ISOLATION FOREST
# ============================================================
X_iso = encoded.reshape(-1, 1)

iso = IsolationForest(contamination=0.02, random_state=42)
iso.fit(X_iso)

pred_iso = iso.predict(X_iso)
anomalies = [1 if x == -1 else 0 for x in pred_iso]

y_true_iso = np.zeros(len(anomalies))
iso_accuracy = accuracy_score(y_true_iso, anomalies)

print("\n==============================")
print("ISOLATION FOREST METRICS")
print("==============================")
print("Accuracy :", round(iso_accuracy * 100, 2), "%")
print("==============================")

# ============================================================
# DEEPLOG DATASET
# ============================================================
if len(encoded) <= SEQ_LEN:
    raise ValueError("Pas assez de logs")

X_seq, y_seq = [], []

for i in range(len(encoded) - SEQ_LEN):
    X_seq.append(encoded[i:i+SEQ_LEN])
    y_seq.append(encoded[i+SEQ_LEN])

X_seq = torch.tensor(X_seq, dtype=torch.long)
y_seq = torch.tensor(y_seq, dtype=torch.long)

# ============================================================
# DEEPLOG MODEL
# ============================================================
class DeepLog(nn.Module):
    def __init__(self, vocab_size):
        super().__init__()
        self.embed = nn.Embedding(vocab_size, 32)
        self.lstm = nn.LSTM(32, 64, batch_first=True)
        self.fc = nn.Linear(64, vocab_size)

    def forward(self, x):
        x = self.embed(x)
        x, _ = self.lstm(x)
        return self.fc(x[:, -1])

model = DeepLog(len(set(encoded)))
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
loss_fn = nn.CrossEntropyLoss()

# ============================================================
# TRAIN + METRICS COLLECTION
# ============================================================
EPOCHS = 10

loss_list = []
acc_list = []

for epoch in range(EPOCHS):

    optimizer.zero_grad()

    outputs = model(X_seq)

    loss = loss_fn(outputs, y_seq)

    loss.backward()

    optimizer.step()

    _, predicted = torch.max(outputs, 1)
    accuracy = (predicted == y_seq).float().mean()

    loss_list.append(loss.item())
    acc_list.append(accuracy.item())

    print("\nEpoch", epoch + 1)
    print("Loss :", round(loss.item(), 4))
    print("Accuracy :", round(accuracy.item() * 100, 2), "%")

print("✅ DeepLog entraîné")

# ============================================================
# MERKLE + SQLITE INSERT
# ============================================================
for i in range(len(logs)):

    log = logs[i]
    template = templates[i]
    cluster_id = cluster_ids[i]
    anomaly = anomalies[i]

    leaf_hash = hashlib.sha256(log.encode()).hexdigest()

    signature = sign_data(leaf_hash)

    tree.add(leaf_hash.encode())

    cursor.execute("""
    INSERT INTO parsed_logs (
        original_log,
        template,
        cluster_id,
        anomaly,
        signature,
        merkle_leaf
    ) VALUES (?, ?, ?, ?, ?, ?)
    """, (
        log,
        template,
        int(cluster_id),
        int(anomaly),
        signature,
        leaf_hash
    ))

conn.commit()

# ============================================================
# MERKLE ROOT
# ============================================================
root_hash = tree.root()
print("\n🌳 ROOT MERKLE:", root_hash)

# ============================================================
# VERIFY SIGNATURES
# ============================================================
cursor.execute("SELECT merkle_leaf, signature FROM parsed_logs LIMIT 5")

print("\nVALIDATION SIGNATURES")
for leaf, sig in cursor.fetchall():
    print("VALID :", verify_signature(leaf, sig))

# ============================================================
# EXPORT
# ============================================================
df = pd.read_sql_query("SELECT * FROM parsed_logs", conn)
df.to_csv("results_bgl.csv", index=False)

torch.save(model.state_dict(), "deeplog_model.pth")

# ============================================================
# 📊 COURBES LOSS + ACCURACY
# ============================================================
epochs = range(1, EPOCHS + 1)

plt.figure()
plt.plot(epochs, loss_list)
plt.title("DeepLog Loss Curve")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid()
plt.show()

plt.figure()
plt.plot(epochs, acc_list)
plt.title("DeepLog Accuracy Curve")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.grid()
plt.show()

plt.figure()
plt.plot(epochs, loss_list, label="Loss")
plt.plot(epochs, acc_list, label="Accuracy")
plt.title("DeepLog Metrics")
plt.xlabel("Epoch")
plt.legend()
plt.grid()
plt.show()

# ============================================================
# CLOSE
# ============================================================
conn.close()

print("\n==============================")
print("✅ PIPELINE COMPLET + COURBES")
print("==============================")
```

```python
# ============================================================
# INSTALLATION
# ============================================================

!pip install -q \
cachetools cryptography pandas numpy scikit-learn torch tqdm matplotlib \
web3==6.15.1 eth-account==0.11.0

# ============================================================
# IMPORTS
# ============================================================

import os
import hashlib
import sqlite3
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import torch
import torch.nn as nn

from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score

from cryptography.hazmat.primitives.asymmetric.ed25519 import Ed25519PrivateKey

from abc import ABC, abstractmethod

# ============================================================
# CONFIG
# ============================================================

BGL_FILE = "/content/drive/MyDrive/BGL_2k.log_structured.csv"
SQLITE_DB = "logs_security.db"
MAX_LOGS = 5000
SEQ_LEN = 10

# ============================================================
# SQLITE INIT
# ============================================================

conn = sqlite3.connect(SQLITE_DB)
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS parsed_logs (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    original_log TEXT,
    template TEXT,
    cluster_id INTEGER,
    anomaly INTEGER,
    signature TEXT,
    merkle_leaf TEXT,
    timestamp DATETIME DEFAULT CURRENT_TIMESTAMP
)
""")

conn.commit()

print("✅ SQLite initialisée")

# ============================================================
# LOAD BGL + INSERT INTO SQLITE
# ============================================================

df_bgl = pd.read_csv(BGL_FILE).head(MAX_LOGS)

rows = [(str(row.iloc[0]),) for _, row in df_bgl.iterrows()]

cursor.executemany("""
INSERT INTO parsed_logs (original_log)
VALUES (?)
""", rows)

conn.commit()

print("✅ BGL inséré dans SQLite")

# ============================================================
# DRAIN IMPLEMENTATION FIXED
# ============================================================

class LogCluster:
    def __init__(self, tokens, cluster_id):
        self.log_template_tokens = tuple(tokens)
        self.cluster_id = cluster_id
        self.size = 1

    def get_template(self):
        return " ".join(self.log_template_tokens)

class Node:
    def __init__(self):
        self.key_to_child_node = {}
        self.cluster_ids = []

class DrainBase(ABC):
    def __init__(self, sim_th=0.4, param_str="<*>"):
        self.sim_th = sim_th
        self.param_str = param_str
        self.root_node = Node()
        self.id_to_cluster = {}
        self.cluster_counter = 0

    def get_tokens(self, content):
        return content.strip().split()

    @abstractmethod
    def tree_search(self, root, tokens, sim_th, include_params): pass

    @abstractmethod
    def add_to_tree(self, root, cluster): pass

    @abstractmethod
    def seq_distance(self, seq1, seq2): pass

    @abstractmethod
    def create_template(self, s1, s2): pass

    def add_log_message(self, content):
        tokens = self.get_tokens(content)

        match = self.tree_search(self.root_node, tokens, self.sim_th, False)

        if match is None:
            self.cluster_counter += 1
            cluster = LogCluster(tokens, self.cluster_counter)
            self.id_to_cluster[self.cluster_counter] = cluster
            self.add_to_tree(self.root_node, cluster)
        else:
            new_template = self.create_template(tokens, match.log_template_tokens)
            match.log_template_tokens = tuple(new_template)
            match.size += 1
            cluster = match

        return cluster, "OK"

# ============================================================
# 🔥 DRAIN (FIX fast_match ERROR HERE)
# ============================================================

class Drain(DrainBase):

    def tree_search(self, root, tokens, sim_th, include_params):

        node = root.key_to_child_node.get(str(len(tokens)))
        if node is None:
            return None

        for t in tokens:
            node = node.key_to_child_node.get(t) or node.key_to_child_node.get(self.param_str)
            if node is None:
                return None

        return self.fast_match(node.cluster_ids, tokens)

    # ✅ FIX ICI (ERREUR CORRIGÉE)
    def fast_match(self, cluster_ids, tokens):

        best = None
        best_sim = -1

        for cid in cluster_ids:
            cluster = self.id_to_cluster.get(cid)
            if cluster is None:
                continue

            sim = sum(a == b for a, b in zip(cluster.log_template_tokens, tokens)) / len(tokens)

            if sim > best_sim:
                best_sim = sim
                best = cluster

        if best_sim >= self.sim_th:
            return best
        return None

    def add_to_tree(self, root, cluster):

        node = root.key_to_child_node.setdefault(str(len(cluster.log_template_tokens)), Node())

        cur = node

        for t in cluster.log_template_tokens:
            if t not in cur.key_to_child_node:
                cur.key_to_child_node[t] = Node()
            cur = cur.key_to_child_node[t]

        cur.cluster_ids.append(cluster.cluster_id)

    def seq_distance(self, s1, s2):
        return sum(a == b for a, b in zip(s1, s2)) / len(s1)

    def create_template(self, s1, s2):
        return [b if a != b else a for a, b in zip(s1, s2)]

# ============================================================
# INIT DRAIN
# ============================================================

parser = Drain()
print("✅ Drain corrigé initialisé")

# ============================================================
# PARSING LOGS
# ============================================================

cursor.execute("SELECT id, original_log FROM parsed_logs")
rows = cursor.fetchall()

templates = []
cluster_ids = []

for log_id, log in rows:
    cluster, _ = parser.add_log_message(log)
    templates.append(cluster.get_template())
    cluster_ids.append(cluster.cluster_id)

print("✅ Parsing terminé")

# ============================================================
# UPDATE SQLITE
# ============================================================

for i, (log_id, _) in enumerate(rows):
    cursor.execute("""
    UPDATE parsed_logs
    SET template = ?, cluster_id = ?
    WHERE id = ?
    """, (templates[i], cluster_ids[i], log_id))

conn.commit()

print("✅ SQLite mise à jour")

# ============================================================
# FIN
# ============================================================

conn.close()

print("🎉 PIPELINE COMPLET OK (BGL + DRAIN FIXED)")
```

```python
# ============================================================
# INSTALLATION
# ============================================================

!pip install -q pandas numpy scikit-learn torch matplotlib

# ============================================================
# IMPORTS
# ============================================================

import sqlite3
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

# ============================================================
# CONFIG
# ============================================================

BGL_FILE = "/content/drive/MyDrive/BGL_2k.log_structured.csv"
DB_NAME = "logs_security.db"
MAX_LOGS = 3000

# ============================================================
# SQLITE
# ============================================================

conn = sqlite3.connect(DB_NAME)
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS logs (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    log TEXT,
    cluster_id INTEGER,
    template TEXT
)
""")

conn.commit()

print("✅ DB ready")

# ============================================================
# LOAD BGL
# ============================================================

df = pd.read_csv(BGL_FILE).head(MAX_LOGS)
logs = df.iloc[:, 0].astype(str).tolist()

cursor.executemany(
    "INSERT INTO logs (log) VALUES (?)",
    [(l,) for l in logs]
)

conn.commit()

print("✅ BGL loaded")

# ============================================================
# SIMPLE DRAIN
# ============================================================

class Cluster:
    def __init__(self, tokens, cid):
        self.tokens = tokens
        self.cid = cid

    def template(self):
        return " ".join(self.tokens)

class Drain:
    def __init__(self):
        self.clusters = {}
        self.counter = 0

    def parse(self, log):

        tokens = log.split()

        best = None
        best_score = 0

        for c in self.clusters.values():
            if len(c.tokens) != len(tokens):
                continue

            score = sum(a == b for a, b in zip(c.tokens, tokens)) / len(tokens)

            if score > best_score:
                best_score = score
                best = c

        if best_score >= 0.6:
            return best

        self.counter += 1
        c = Cluster(tokens, self.counter)
        self.clusters[self.counter] = c
        return c

parser = Drain()

# ============================================================
# PARSING
# ============================================================

cursor.execute("SELECT id, log FROM logs")
rows = cursor.fetchall()

cluster_ids = []
templates = []

for log_id, log in rows:
    c = parser.parse(log)
    cluster_ids.append(c.cid)
    templates.append(c.template())

print("✅ Parsing done")

# ============================================================
# UPDATE DB
# ============================================================

for i, (log_id, _) in enumerate(rows):
    cursor.execute("""
    UPDATE logs
    SET cluster_id = ?, template = ?
    WHERE id = ?
    """, (cluster_ids[i], templates[i], log_id))

conn.commit()

print("✅ DB updated")

# ============================================================
# ML MODEL
# ============================================================

X = torch.tensor(cluster_ids, dtype=torch.float32).view(-1, 1)

counts = pd.Series(cluster_ids).value_counts()
y = np.array([1 if counts[c] < 3 else 0 for c in cluster_ids])
y = torch.tensor(y, dtype=torch.float32).view(-1, 1)

model = nn.Sequential(
    nn.Linear(1, 16),
    nn.ReLU(),
    nn.Linear(16, 1),
    nn.Sigmoid()
)

loss_fn = nn.BCELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

# ============================================================
# TRAINING (FIX + METRICS STORAGE)
# ============================================================

loss_list = []
acc_list = []

EPOCHS = 25

for epoch in range(EPOCHS):

    optimizer.zero_grad()

    out = model(X)

    loss = loss_fn(out, y)

    loss.backward()
    optimizer.step()

    preds = (out > 0.5).float()
    acc = (preds == y).float().mean()

    loss_list.append(loss.item())
    acc_list.append(acc.item())

    print(f"Epoch {epoch+1} | Loss: {loss.item():.4f} | Acc: {acc.item():.4f}")

# ============================================================
# FINAL RESULTS
# ============================================================

final_acc = (model(X) > 0.5).float().eq(y).float().mean().item()

print("\n======================")
print("FINAL RESULTS")
print("======================")
print("Accuracy:", round(final_acc, 4))
print("Loss:", round(loss_list[-1], 4))

# ============================================================
# PLOTS (SIDE-BY-SIDE FIXED)
# ============================================================

epochs_range = range(1, len(loss_list) + 1)

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# LOSS
ax[0].plot(epochs_range, loss_list)
ax[0].set_title("Loss Curve")
ax[0].set_xlabel("Epoch")
ax[0].set_ylabel("Loss")
ax[0].grid(True)

# ACCURACY
ax[1].plot(epochs_range, acc_list)
ax[1].set_title("Accuracy Curve")
ax[1].set_xlabel("Epoch")
ax[1].set_ylabel("Accuracy")
ax[1].grid(True)

plt.tight_layout()
plt.show()

# ============================================================
# CLOSE
# ============================================================

conn.close()

print("🎉 PIPELINE COMPLETE")
```
