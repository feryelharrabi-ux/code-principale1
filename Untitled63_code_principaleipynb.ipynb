```python
# ============================================================
# 📘 LOG SECURITY & INTEGRITY PIPELINE (RESEARCH VERSION)
# ============================================================

!pip install -q drain3 cryptography pandas numpy scikit-learn torch tqdm matplotlib

# ============================================================
# IMPORTS
# ============================================================

import hashlib
import sqlite3
import numpy as np
import pandas as pd
import time
import matplotlib.pyplot as plt
from tqdm import tqdm

import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset

from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import IsolationForest
from sklearn.metrics import precision_score, recall_score, f1_score

from drain3 import TemplateMiner
from drain3.template_miner_config import TemplateMinerConfig

from cryptography.hazmat.primitives.asymmetric.ed25519 import Ed25519PrivateKey


# ============================================================
# CONFIG
# ============================================================

BGL_FILE = "/content/drive/MyDrive/BGL_2k.log_structured.csv"
MAX_LOGS = 3000
SEQ_LEN = 10
EPOCHS = 6
BATCH_SIZE = 64


# ============================================================
# SQLITE
# ============================================================

conn = sqlite3.connect("logs_security.db")
cursor = conn.cursor()

cursor.execute("DROP TABLE IF EXISTS parsed_logs")

cursor.execute("""
CREATE TABLE parsed_logs (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    original_log TEXT,
    template TEXT,
    cluster_id INTEGER,
    anomaly INTEGER,
    signature TEXT,
    merkle_leaf TEXT,
    timestamp TEXT
)
""")
conn.commit()


# ============================================================
# LOAD DATA
# ============================================================

df = pd.read_csv(BGL_FILE).head(MAX_LOGS)
logs = df.iloc[:, 0].astype(str).tolist()

split_logs = int(0.8 * len(logs))
train_logs = logs[:split_logs]
test_logs = logs[split_logs:]


# ============================================================
# DRAIN3 (NO LEAKAGE FIX)
# ============================================================

config = TemplateMinerConfig()
config.drain_depth = 4
config.drain_sim_th = 0.4

miner = TemplateMiner(config=config)

templates = []
cluster_ids = []

# TRAIN ONLY
for log in train_logs:
    res = miner.add_log_message(log)
    templates.append(res["template_mined"])
    cluster_ids.append(res["cluster_id"])

# TEST APPLY SAME MODEL
for log in test_logs:
    res = miner.add_log_message(log)
    templates.append(res["template_mined"])
    cluster_ids.append(res["cluster_id"])


# ============================================================
# ENCODING
# ============================================================

encoder = LabelEncoder()
encoded = encoder.fit_transform(templates)


# ============================================================
# SEQUENCES (DeepLog)
# ============================================================

X, y = [], []

for i in range(len(encoded) - SEQ_LEN):
    X.append(encoded[i:i+SEQ_LEN])
    y.append(encoded[i+SEQ_LEN])

X = torch.tensor(X, dtype=torch.long)
y = torch.tensor(y, dtype=torch.long)

split = int(0.8 * len(X))

X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]


# ============================================================
# ISOLATION FOREST (CLEAN)
# ============================================================

X_iso = encoded.reshape(-1, 1)

iso_split = int(0.8 * len(X_iso))

iso = IsolationForest(contamination=0.02, random_state=42)
iso.fit(X_iso[:iso_split])

iso_pred = iso.predict(X_iso[iso_split:])
anomalies_iso = (iso_pred == -1).astype(int)


# ============================================================
# DATA LOADER
# ============================================================

train_loader = DataLoader(
    TensorDataset(X_train, y_train),
    batch_size=BATCH_SIZE,
    shuffle=True
)


# ============================================================
# DEEPLOG MODEL
# ============================================================

class DeepLog(nn.Module):
    def __init__(self, vocab):
        super().__init__()
        self.emb = nn.Embedding(vocab, 32)
        self.lstm = nn.LSTM(32, 64, batch_first=True)
        self.fc = nn.Linear(64, vocab)

    def forward(self, x):
        x = self.emb(x)
        x, _ = self.lstm(x)
        return self.fc(x[:, -1])

model = DeepLog(len(set(encoded)))

optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
loss_fn = nn.CrossEntropyLoss()

loss_list = []


# ============================================================
# TRAINING (STABLE)
# ============================================================

for epoch in range(EPOCHS):
    total_loss = 0

    for xb, yb in train_loader:
        optimizer.zero_grad()
        out = model(xb)
        loss = loss_fn(out, yb)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

    loss_list.append(total_loss)
    print(f"Epoch {epoch+1} | Loss {total_loss:.4f}")


# ============================================================
# ANOMALY SCORING
# ============================================================

with torch.no_grad():
    out = model(X_test)
    probs = torch.softmax(out, dim=1)

    anomaly_score = -torch.max(probs, dim=1).values
    threshold = anomaly_score.mean() + anomaly_score.std()

    anomalies_pred = anomaly_score > threshold


# ============================================================
# METRICS
# ============================================================

preds = torch.argmax(out, dim=1)

print("\nPrecision:", precision_score(y_test, preds, average="macro"))
print("Recall:", recall_score(y_test, preds, average="macro"))
print("F1:", f1_score(y_test, preds, average="macro"))


# ============================================================
# CRYPTO LAYER
# ============================================================

private_key = Ed25519PrivateKey.generate()
public_key = private_key.public_key()

def sign_data(x):
    return private_key.sign(x.encode()).hex()


# ============================================================
# MERKLE TREE
# ============================================================

class MerkleTree:
    def __init__(self):
        self.leaves = []

    def add(self, data):
        self.leaves.append(hashlib.sha256(data).hexdigest())

    def root(self):
        nodes = self.leaves[:]

        while len(nodes) > 1:
            temp = []
            for i in range(0, len(nodes), 2):
                left = nodes[i]
                right = nodes[i+1] if i+1 < len(nodes) else left

                temp.append(hashlib.sha256((left + right).encode()).hexdigest())

            nodes = temp

        return nodes[0] if nodes else None


tree = MerkleTree()


# ============================================================
# BLOCKCHAIN SIMULATION
# ============================================================

class LogBlockchain:
    def __init__(self):
        self.chain = []
        self.prev_hash = "0"

    def add_block(self, log, merkle_root):
        block = {
            "index": len(self.chain),
            "timestamp": time.time(),
            "log": log,
            "prev_hash": self.prev_hash,
            "merkle_root": merkle_root
        }

        block["hash"] = hashlib.sha256(str(block).encode()).hexdigest()

        self.prev_hash = block["hash"]
        self.chain.append(block)

        return block


blockchain = LogBlockchain()


def verify_chain(chain):
    for i in range(1, len(chain)):
        if chain[i]["prev_hash"] != chain[i-1]["hash"]:
            return False
    return True


# ============================================================
# STORAGE + BLOCKCHAIN
# ============================================================

for i, log in enumerate(logs):

    leaf = hashlib.sha256(log.encode()).hexdigest()
    sig = sign_data(leaf)

    tree.add(leaf.encode())
    merkle_root = tree.root()

    block = blockchain.add_block(log, merkle_root)

    cursor.execute("""
    INSERT INTO parsed_logs VALUES (NULL,?,?,?,?,?,?,datetime('now'))
    """, (
        log,
        templates[i],
        cluster_ids[i],
        int(anomalies_pred[i % len(anomalies_pred)]),
        sig,
        leaf
    ))

conn.commit()


# ============================================================
# OUTPUT FINAL
# ============================================================

print("\nMERKLE ROOT:", tree.root())
print("BLOCKCHAIN VALID:", verify_chain(blockchain.chain))
print("BLOCKS:", len(blockchain.chain))


# ============================================================
# SAVE MODEL
# ============================================================

torch.save(model.state_dict(), "deeplog.pth")

conn.close()

print("🎉 RESEARCH PIPELINE COMPLETE")
```
